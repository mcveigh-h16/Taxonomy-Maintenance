```python
import pandas as pd
import numpy as np
import sys
```

```python
MBList = (r'MBList2025.xlsx')
MBist_df = pd.read_excel(MBList, index_col=None)
MBList_df = MBist_df.rename(columns={"ID": "id", "Taxon name": "name"}, inplace=True)
MBist_df.set_index('id', inplace=True)
MBist_df
```

```python
Mycobank_api = (r'mycobank449.xlsx')
Mycobank_api_df = pd.read_excel(Mycobank_api, index_col=None)
Mycobank_api_df.set_index('id', inplace=True)
Mycobank_api_df
```

```python
combined = pd.merge(Mycobank_api_df, right=MBist_df, left_on='id', right_on='id', how='outer')
#combined.drop_duplicates(subset=['protein_accession'], inplace=True)
#combined.drop(columns=['orientation', 'location', 'gene_symbol', 'protein_name'], inplace=True)
#combined.to_csv(r'table_combined', sep ='\t', index = True) 
combined
```
