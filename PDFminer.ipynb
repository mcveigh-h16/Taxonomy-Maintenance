```python
def extract_text_from_pdf(pdf_path, start_page, end_page, regex_pattern):
    """Extracts text from a specified page range of a PDF using regex."""

    text_list = []

    with open(pdf_path, 'rb') as pdf_file:
        reader = PyPDF2.PdfReader(pdf_file)
        num_pages = len(reader.pages)

        # Check if the page range is valid
        if start_page < 1 or end_page > num_pages or start_page > end_page:
            raise ValueError("Invalid page range")

        # Extract text from the specified pages
        for page_num in range(start_page - 1, end_page):
            page = reader.pages[page_num]
            text = page.extract_text()

            # Use regex to find the specific text
            matches = re.findall(regex_pattern, text)
            text_list.extend(matches)

    return text_list

# Example usage
pdf_path = '2024__The2024OutlineofFungiandfungus-liketaxa.pdf'
start_page = 41
end_page = 332
regex_pattern = r'Phylum'  # Example: find dates in the format DD/MM/YYYY

# Extract text and create a DataFrame
extracted_text = extract_text_from_pdf(pdf_path, start_page, end_page, regex_pattern)
df = pd.DataFrame({'Text': extracted_text})

print(df)
```

```python
def extract_text_from_page_span(pdf_path, start_page, end_page):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page_num in range(start_page - 1, end_page): 
            page = reader.pages[page_num]
            text += page.extract_text()
    return text

pdf_path = "2024__The2024OutlineofFungiandfungus-liketaxa.pdf"
start_page = 41
end_page = 50
extracted_text = extract_text_from_page_span(pdf_path, start_page, end_page)

extracted_text = re.sub(r",\s*\n", " ", extracted_text)

text_lines = extracted_text.splitlines()

for line in text_lines:
    print(line)
```

```python
def flip_first_two_words(text, match_list):
    words = text.split()
    if len(words) >= 2 and words[0] in match_list:
        words[0], words[1] = words[1], words[0]
    return ' '.join(words)

match_list = ['Phylum', 'Class', 'Subphylum']

df['flipped_text'] = df['Text'].apply(lambda x: flip_first_two_words(x, match_list))
```

```python
def join_indented_lines(text):
    lines = text.splitlines()
    result = []
    previous_line = None

    for line in lines:
        #if line.startswith((' ', '\t')):  # Check for indentation (spaces or tabs)
        if line.startswith(' '):  # Check for indentation (spaces or tabs)
            if previous_line is not None:
                result[-1] = result[-1] + ' ' + line.lstrip()  # Append to previous
            else:
                result.append(line.lstrip()) #if the first line is indented, append it directly
        else:
            result.append(line)
            previous_line = line

    return '\n'.join(result)

```
